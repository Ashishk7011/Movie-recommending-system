```python
import pandas as pd
import numpy as np
```

```python
films= pd.read_csv('tmdb_5000_movies.csv')
```

```python
credits= pd.read_csv('tmdb_5000_credits.csv')
```

```python
films= films.merge(credits, on = 'title')
```

```python
films= films[['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']]
```

```python
films.dropna(inplace=True)
```

```python
def change(item):
    a=[]
    for i in eval(item):
        a.append(i['name'])
    return a
```

```python
films['genres']= films['genres'].apply(change)
```

```python
films
```

```python
def change2(item):
    a=[]
    count=0
    for i in eval(item):
        if count != 3:
            a.append(i['name'])
            count += 1
        else:
            break
    return a
```

```python
films.cast= films['cast'].apply(change2)
```

```python
films.keywords= films.keywords.apply(change)
```

```python
def finddirector(item):
    a=[]
    for i in eval(item):
        if i['job'] == 'Director':
            a.append(i['name'])
            break
    return a
```

```python
films.crew= films.crew.apply(finddirector)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors= cv.fit_transform(new_data['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemmer(text):
    a= []
    for i in text.split():
        a.append(ps.stem(i))
        
    return " ".join(a)    
```

```python
def recommend(search):
    for item in new_data.title:
        if item.lower()==search.lower():
            
            movie_index= new_data[new_data.title==item].index[0]

            similar= similarity[movie_index]
            movies_list= sorted(list(enumerate(similar)), reverse=True, key = lambda x: x[1])[1:6]

            for i in movies_list:
                print(new_data.iloc[i[0]].title)
                
    if any(films.crew.str.contains(search, case=False, na=False)) and any(films.cast.str.contains(search, case=False, na=False)):
        
        crew= films.crew[films[films.crew.str.contains(r'\b{}\b'.format(search), case=False, na=False)].index[0:5]]
        title= films.title[films[films.crew.str.contains(r'\b{}\b'.format(search), case=False, na=False)].index[0:5]]
        crew_df= pd.DataFrame({'crew':crew, 'title':title})
        
        cast= films.cast[films[films.cast.str.contains(r'\b{}\b'.format(search), case=False, na=False)].index[0:5]]
        title2= films.title[films[films.cast.str.contains(r'\b{}\b'.format(search), case=False, na=False)].index[0:5]]
        cast_df= pd.DataFrame({'cast':cast, 'title':title2})
        
        return crew_df, cast_df

    if any(films.crew.str.contains(search, case=False, na=False)):
        crew= films.crew[films[films.crew.str.contains(r'\b{}\b'.format(search), case=False, na=False)].index[0:5]]
        title= films.title[films[films.crew.str.contains(r'\b{}\b'.format(search), case=False, na=False)].index[0:5]]
        return pd.DataFrame({'crew':crew, 'title':title})

    elif any(films.cast.str.contains(search, case=False, na=False)):
        cast= films.cast[films[films.cast.str.contains(r'\b{}\b'.format(search), case=False, na=False)].index[0:5]]
        title= films.title[films[films.cast.str.contains(r'\b{}\b'.format(search), case=False, na=False)].index[0:5]]
        return pd.DataFrame({'cast':cast, 'title':title})

    elif any(films.genres.str.contains(search, case=False, na=False)):
        genres= films.genres[films[films.genres.str.contains(r'\b{}\b'.format(search), case=False, na=False)].index[0:5]]
        title= films.title[films[films.genres.str.contains(r'\b{}\b'.format(search), case=False, na=False)].index[0:5]]
        return pd.DataFrame({'genres':genres, 'title':title})
```
